```python
###STANDARDIZATION
```

```python
#it make the std dev 1 for the dataset
def standardize(x):
    val = (x-x.mean()) / x.std()
    return val
```

```python
x_new = standardize(x)
```

```python
x_new.std()
```

```python
wf = pd.read_csv('wisc_bc_data.csv')
```

```python
wf
```

```python
# separate the input and output variables

x=wf.drop(['id','diagnosis'],axis=1) # features
y=wf['diagnosis'] # outcome
```

```python
x.shape
```

```python
from sklearn.feature_selection import SelectKBest, chi2
```

```python
skf = SelectKBest(score_func=chi2, k=5)
```

```python
skf.fit(x,y)
```

```python
skf.get_support()
```

```python
x_new = x.iloc[:, skf.get_support()]
x_new
```
