```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as mp
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
x=np.array([11,22,33,44,55,66,77,88,99,10])
y=np.array([11,20,30,14,55,16,17,18,19,20])
z=np.array([29,24,26,25,27,78,27,26,29,31])
```

```python
#Line plot
```

```python
mp.plot(x,y);
```

```python
mp.plot(x,z)
```

```python
mp.title('Line Plot')
mp.xlabel('First value')
mp.ylabel('Second value')
mp.plot(x,y,marker='o',color='blue',mfc='r',ls=':',ms=5)


mp.grid()
#ls=line style,ms=marker size
#marker= >,<,o,^,*,H
```

```python
mp.title('Line Plot')
mp.xlabel('First value')
mp.ylabel('Second value')
mp.plot(x,z,marker='*',color='blue',mfc='r',ls='--',ms=8)
#ls; supported values are '-', '--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted'
```

```python
#combine 2 graphs
mp.title('Line Plot')
mp.xlabel('First value')
mp.ylabel('Second value')
mp.plot(x,y,marker='o',color='blue',mfc='r',ls=':',ms=5)
mp.plot(x,z,marker='*',color='orange',mfc='r',ls='--',ms=8)
mp.grid()
```

```python
mp.figure(figsize=(10,10))
#subplot=(height,width,position)
mp.subplot(2,2,1)
mp.title('Line Plot')
mp.xlabel('First value')
mp.ylabel('Second value')
mp.plot(x,y,marker='o',color='blue',mfc='r',ls=':',ms=5)


mp.grid()

mp.subplot(2,2,2)
mp.title('Line Plot')
mp.xlabel('First value')
mp.ylabel('Second value')

mp.plot(x,z,marker='*',color='orange',mfc='r',ls='--',ms=8)
mp.grid()
```

```python
students = [56,65,46,51,61]
courses=['Java','Python','R','Kotlin','Dart']
```

```python
mp.title('Course Enrollement')
mp.xlabel('Courses')
mp.ylabel('No.of Students')
mp.bar(courses,students,color='violet',edgecolor='black',alpha=0.2)
#alpha=opacity(makes shade lighter it lies between 0-1)
#0=white
#1=Darker shade of color
```

```python
mp.scatter(x,y,marker="*",color="black")
mp.grid()
```

```python
mp.title('Pie Chart')
mp.pie(students, labels=courses, autopct='%.2f%%',shadow=True,radius=0.8,
       labeldistance=1.2,colors=['pink','red','green','yellow','violet'],
       explode=[0,0,0.2,0,0]);
```

```python
mp.figure(figsize=(10,10))
#subplot=(height,width,position)
mp.subplot(2,2,1)
mp.title('Line Plot')
mp.xlabel('First value')
mp.ylabel('Second value')
mp.plot(x,y,marker='o',color='blue',mfc='r',ls=':',ms=5)
mp.grid()

mp.subplot(2,2,2)
mp.title('Line Plot')
mp.xlabel('First value')
mp.ylabel('Second value')
mp.plot(x,z,marker='*',color='orange',mfc='r',ls='--',ms=8)
mp.grid()

mp.subplot(2,2,3)
mp.title('Course Enrollement')
mp.xlabel('Courses')
mp.ylabel('No.of Students')
mp.bar(courses,students,color='violet',edgecolor='black',alpha=0.2)

mp.subplot(2,2,4)
mp.title('Pie Chart')
mp.pie(students, labels=courses, autopct='%.2f%%',shadow=True,radius=0.8,
       labeldistance=1.2,colors=['pink','red','green','yellow','violet'],
       explode=[0,0,0.2,0,0])
mp.savefig('My plot')
```

```python
marks=[67,87,85,45,68,89,30,68,69,120]
mp.boxplot(marks);
```

```python
q3=np.percentile(marks,75)
q1=np.percentile(marks,25)
q3,q1
```

```python
#iqr interquartile range
iqr=q3-q1
iqr
```

```python
#lower bound
lb=q1 - 1.5 * iqr

#upper bound
up=q3 + 1.5 * iqr
```

```python
lb,up
```

```python
tips=sns.load_dataset('tips')
tips
```

```python
import matplotlib as mpl
mpl.rcParams.update(mpl.rcParamsDefault)
```

```python
sns.jointplot(x='tip',y='total_bill',data=tips,hue='time')
mp.show()
```

```python
#histogram
mp.xlabel('Tip Amount')
mp.ylabel('Frequency')
sns.histplot(tips['tip'],kde=True)
mp.show()
```

```python
mp.title('Outliers in Tip')
mp.xlabel('Day')
mp.ylabel('tip')
sns.boxenplot(x='day',y='tip',data=tips)

mp.show()
```

```python
sns.pairplot(tips,hue='smoker')
mp.show()
```
