```python
x = np.array([11,45,23,17,21,15,7,33])
```

```python
def normalize(a):
    norm = (a - a.min()) / (a.max() - a.min())
    return norm
```

```python
normalize(x)
```

```python
nx = normalize(x)
```

```python
plt.figure(figsize=(16,9))
plt.subplot(1,2,1)
plt.title('Actual Data')
plt.plot(x)
plt.subplot(1,2,2)
plt.title('Normalize data')
plt.plot(nx)
```

```python
ds = pd.read_csv('Social_Network_Ads.csv')
```

```python
ds
```

```python
x = ds[['Age','EstimatedSalary']]
y = ds['Purchased']
```

```python
plt.xlabel('Age')
plt.ylabel('Salary')
plt.grid()
plt.scatter(x['Age'], x['EstimatedSalary'], c=y);
```

```python
x.describe()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
x_scaled = scaler.fit_transform(x)
```

```python
type(x_scaled)
```

```python
x_new = pd.DataFrame(x_scaled, columns=x.columns)
```

```python
x_new
```

```python
x_new.describe()
```

```python
plt.figure(figsize=(16,9))
plt.subplot(1,2,1)
plt.title('Actual Data')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.grid()
plt.scatter(x_new['Age'], x_new['EstimatedSalary'], c=y)

plt.subplot(1,2,2)
plt.title('Normalize data')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.grid()
plt.scatter(x_new['Age'], x_new['EstimatedSalary'], c=y)
```
